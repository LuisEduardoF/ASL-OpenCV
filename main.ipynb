```python
import pandas as pd

df = pd.read_csv("data/asl_dataset.csv", index_col=0)
```

```python
from sklearn.model_selection import train_test_split

X = df.drop("class", axis=1)
y = df["class"]

# Split the data into training (70%) and testing (30%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
from models import AlphabetModel

asl = AlphabetModel(X_train, y_train, X_test, y_test)
```

```python
print(asl)
```

```python
epochs, losses = asl.fit()
```

```python
from matplotlib import pyplot as plt
# Sample data
x = range(0, epochs) # X-axis values (e.g., time or data points)
y = losses  # Y-axis values (e.g., measurements or values)

# Create a line plot
plt.plot(x, y, label='Loss')

# Add labels and a title
plt.xlabel('Epochs')
plt.ylabel('Losses')
plt.title('Loss/Epoch Plot')
```

```python
asl.score()
```
